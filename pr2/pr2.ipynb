### Задание 1
Построить график функции.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def y(arg): return -0.01 * np.power(arg, 2) - 0.8 * arg * np.sin(1.2 * arg) + 2
```

```python
x_space = np.linspace(start=-10, stop=10, num=400)
y_space = y(x_space)
y_space
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_space, y_space, color='orange', linewidth=3,
        label=r'$y=-0.01 x^2 - 0.8x * \sin{(1.2x)} + 2$')
ax.legend()
ax.set_title('Заголовок графика', fontsize=16)
ax.set_xlabel(r'$x$', fontsize=14)
ax.set_ylabel(r'$y$', fontsize=14)
ax.grid()
plt.savefig('okaivd_pr2_1.jpg', dpi=150, bbox_inches='tight')
plt.show()
```
