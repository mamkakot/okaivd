### Задание 1
Сгенерировать массив из 100 элементов - переменная x1 в интервале [-3, 7], значения равномерно расположены в интервале (между 2мя точками одинаковое расстояние).

```python
import numpy as np
x1 = np.linspace(-3, 7, 100)
y1 = np.sin(x1)
# некоторые строки с выводом, как ниже, я закомментил, чтобы оно не гадило в вывод, если интересно, что там в переменной, можно убрать комменты
# y1
```

### Задание 2
Сгенерировать массив из 100 элементов - переменная x2 в интервале [-3, 7], значения сгенерированы равномерно случайно.

```python
x2 = np.random.uniform(low=-3, high=7, size=100)
y2 = np.sin(x2)
# y2
```

### Задание 3
Сгенерировать массив - переменная x3 в интервале [-3, 7], значения равномерно расположены в интервале с шагом 0.05
Переменная noise - это имитация шума (ошибок в измерениях). В каждом измерении (для каждой вычисленной точки y3 генерируется свое значение шума) шум представлен случайным, нормально распределенным значением с параметрами N(0, 0.15).

```python
x3 = np.arange(-3, 7, 0.05)
y3 = np.sin(x3) + np.random.uniform(0, .15, x3.shape)
# y3
```

### Задание 4
Сгенерировать 2 массива по 100 элементов - переменные x и y в интервале [-3, 7], значения сгенерированы равномерно случайно.

```python
x = np.random.uniform(low=-3, high=7, size=100)
y = np.random.uniform(low=-3, high=7, size=100)
z = np.power(x, 2) + np.power(y, 2) + np.sin(x) + np.cos(y)
# z
```

# Графики полученных функций
